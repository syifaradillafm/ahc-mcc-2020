```
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cmath as math
import sys

#loading the dataset
dataset = pd.read_excel('/Data AHC - 3.xlsx')
X = dataset.iloc[:,[1,4,5]].values
y_set = dataset.iloc[:,[0]].values
```

```
# Semua Cluster
X = dataset.iloc[:, [4,5,1]].values

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'complete'))
plt.title('Dendrogram')
plt.xlabel('Data Cluster')
plt.ylabel('Jarak Euclidean')
plt.show()

from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'complete')
y_hc = hc.fit_predict(X)
```

```
# Cluster 1 - Masa Studi 3.5 Tahun

dataset_c1 = dataset[(dataset.IPK>=3.8) & (dataset.Kerja==0) & (dataset.Organisasi==0)]
X = dataset_c1.iloc[:, [0,2]].values

# Menggunakan dendrogram untuk menentukan angka cluster yang tepat
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'complete'))
plt.title('Dendrogram')
plt.xlabel('Cluster 1')
plt.ylabel('Jarak Euclidean')
plt.show()

# Menjalankan Hierarchical Clustering ke dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 1, affinity = 'euclidean', linkage = 'complete')
y_hc = hc.fit_predict(X)
```
